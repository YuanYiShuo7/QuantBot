```python
import easyquotation

# 初始化接口（推荐使用新浪免费源）
quotation = easyquotation.use('sina')

# 获取单只股票实时数据
stock_code = '000001'  # 平安银行
data = quotation.real(stock_code)
print(data)

# 获取多只股票实时数据
stock_codes = ['000001', '000858', '600519']  # 平安银行、五粮液、贵州茅台
multi_data = quotation.real(stock_codes)
print(multi_data)
```

```python
import akshare as ak

def test_weekly_data():
    """测试周线和月线数据"""
    print("=" * 50)
    print("周线月线数据测试")
    print("=" * 50)
    
    # 测试周线数据
    try:
        stock_zh_a_hist_weekly_df = ak.index_zh_a_hist(symbol="000001", period="weekly", 
                                                      start_date="20230601", end_date="20230607", 
                                                      )
        print("周线数据字段:")
        print(f"列名: {list(stock_zh_a_hist_weekly_df.columns)}")
        print(f"数据样例:\n{stock_zh_a_hist_weekly_df.head(2)}")
        print(f"数据形状: {stock_zh_a_hist_weekly_df.shape}")
        print()
    except Exception as e:
        print(f"周线数据获取失败: {e}")
    
if __name__ == "__main__":
    test_weekly_data()
```
