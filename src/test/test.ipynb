```python
import akshare as ak
import pandas as pd

print(ak.__version__)

import akshare as ak
import pandas as pd
from datetime import datetime

def get_stock_history_corrected(stock_code, period="daily", start_date="20200101", end_date="20231231"):
    """
    修正版的股票历史数据获取函数
    """
    try:
        print(f"正在获取 {stock_code} 的数据...")
        print(f"时间范围: {start_date} 到 {end_date}")
        
        # 方法1：使用新的接口（推荐）
        try:
            # 新的接口格式
            stock_zh_a_hist_df = ak.stock_zh_a_hist(
                symbol=stock_code,  # 注意：现在不需要加市场后缀了
                period=period,
                start_date=start_date,
                end_date=end_date,
                adjust="hfq"  # 添加复权参数
            )
        except Exception as e1:
            print(f"新接口失败: {e1}")
        
        if stock_zh_a_hist_df is not None and len(stock_zh_a_hist_df) > 0:
            print(f"成功获取 {len(stock_zh_a_hist_df)} 条数据")
            return stock_zh_a_hist_df
        else:
            print("获取到的数据为空")
            return None
            
    except Exception as e:
        print(f"获取数据失败: {e}")
        return None

# 测试修正后的函数
print("=" * 50)
data = get_stock_history_corrected("000001", start_date="20230101")
if data is not None:
    print("数据预览:")
    print(data.head())
else:
    print("没有获取到数据")


# 获取A股历史行情数据
def get_stock_history(stock_code, period="daily", start_date="20200101", end_date="20231231"):
    """
    获取股票历史数据
    stock_code: 股票代码，如 '000001'（平安银行）
    period: 周期，daily-日线, weekly-周线, monthly-月线
    """
    try:
        # 注意：需要添加市场后缀，sz-深交所, sh-上交所
        if stock_code.startswith('0') or stock_code.startswith('3'):
            code_with_market = f"{stock_code}.sz"
        else:
            code_with_market = f"{stock_code}.sh"
        
        stock_zh_a_hist_df = ak.stock_zh_a_hist(
            symbol=code_with_market,
            period=period,
            start_date=start_date,
            end_date=end_date,
        )
        return stock_zh_a_hist_df
    except Exception as e:
        print(f"获取数据失败: {e}")
        return None

data = get_stock_history("000001", start_date="20230101")
print(data.head())
```
